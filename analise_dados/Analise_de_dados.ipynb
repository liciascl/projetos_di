```python
# Histograma com a quantidade de ataques por ano

# Gerando um histograma
df.iyear.hist(bins = 90)
plt.xlabel("Ano")
plt.ylabel("Ataques Registrados")
plt.title("Ataques Terroristas por ano")
plt.show()
```

```python
# Quais são os alvos mais comuns para ataques terroristas?

# Definindo a quantidade
labels = df.targtype3_txt.value_counts().index
num = len(df.target3.value_counts().index)

# Gráfico de Pizza
fatias, texto = plt.pie(df.targtype3_txt.value_counts(), startangle = 90)
plt.legend(fatias, labels, bbox_to_anchor = (1.05,1))
plt.title("Alvos Preferidos")
plt.show()
```

```python
# Quais são os tipos de ataques?

# Definindo a quantidade
labels = df.attacktype3_txt.value_counts().index
num = len(df.attacktype3.value_counts().index)

# Gráfico de Pizza
fatias, texto = plt.pie(df.attacktype3_txt.value_counts(), startangle = 90)
plt.legend(fatias, labels, bbox_to_anchor = (1.05,1))
plt.title("Tipos de Ataques")
plt.show()

```

```python
# Quais são as Regiões mais atacadas?

# Definindo a quantidade
labels = df.region_txt.value_counts().index
num = len(df.region.value_counts().index)

# Gráfico de Pizza
fatias, texto = plt.pie(df.region_txt.value_counts(), startangle = 90)
plt.legend(fatias, labels, bbox_to_anchor = (1.05,1))
plt.title("Alvos")
plt.show()

```

```python
# Quais são as Cidades mais atacadas?

# Definindo a quantidade
labels = df.country_txt.value_counts().index
num = len(df.country.value_counts().index)

# Gráfico de Pizza
fatias, texto = plt.pie(df.country.value_counts(), startangle = 90)
plt.legend(fatias, labels, bbox_to_anchor = (1.05,1))
plt.title("Cidades Alvo")
plt.show()
```

```python
# Agrupando os dados
df3 = pd.crosstab(df.weaptype1_txt, 
                  df.success).apply(lambda r: r/r.sum(), axis = 1)
# Definindo a quantidade
num = len(df.weaptype1_txt.value_counts().index)

# Gráfico de Barras (Stacked)
ax1 = df3.plot(kind = "bar", stacked = True, title = "Tipo de Arma x Sucesso da operação")
lines, labels = ax1.get_legend_handles_labels()
ax1.legend(lines,["Falhou", "Sucesso"], loc = 'best')
```
